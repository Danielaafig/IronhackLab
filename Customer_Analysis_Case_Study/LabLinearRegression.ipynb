```python
#X-y split 
X = numerical.drop('total_claim_amount', axis=1) #(y is the target variable, which is the total claim amount)
y = numerical.total_claim_amount
                                 
X
```

```python
std_scaler=StandardScaler().fit(X_train)   ##. finding the parameters ( mean, variance from the training set )

X_train_scaled=std_scaler.transform(X_train)            #Standardize the data (after the data split).
X_train.shape
```

```python
X_train_const_scaled = sm.add_constant(X_train_scaled) # adding a constant

model = sm.OLS(y_train, X_train_const_scaled).fit()
predictions_train = model.predict(X_train_const_scaled) 

X_test_const_scaled = sm.add_constant(X_test_scaled) # adding a constant (Beta)
predictions_test = model.predict(X_test_const_scaled) 
print_model = model.summary()                                               #Apply linear regression.
print(print_model)
```
